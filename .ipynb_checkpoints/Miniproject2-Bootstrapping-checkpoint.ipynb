```python
import numpy as np 
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt 
import seaborn; seaborn.set()
```

```python
root = '/Users/caichengyun/Coding/NYU/FRE-6103/Data'
cols = [i for i in range(1, 7)]
df = pd.read_excel(root + '/US Bond Quotes 091621.xlsx', skiprows=4, usecols=cols)
df.drop(df.index[-1], inplace=True)

df
```

```python
df['DAYS TO MATURITY'] = -pd.to_datetime(df['MATURITY']).rsub(pd.Timestamp('now').floor('d')).dt.days

df.head()
```

```python
ax = df.plot(
    x="DAYS TO MATURITY", y="ASKED YIELD",
    kind="line", linestyle="", marker="."
)
```

```python
def f_model(t, a1, a2, a3, a4, a5, a6, a7):
    return a1+a2*t+a3*(t**2)+a4*np.exp(a5*t)+a6*np.log(t-a7)
```

```python
df["MODEL"] = f_model(df["DAYS TO MATURITY"], 0.0001, 0.0001, 0.0001, 0.0001, 0.0001, 0.0001, 0.0001)
df.head(8)
```

```python
ax = df.plot(
    x="DAYS TO MATURITY", y="ASKED YIELD",
    kind="line", linestyle="", marker="."
)
ax = df.plot(
    x="DAYS TO MATURITY", y="MODEL",
    ax=ax, linewidth=1
)
```

```python
from scipy.optimize import curve_fit

popt, pcov = curve_fit(
    f=f_model,       # model function
    xdata=df["DAYS TO MATURITY"],   # x data
    ydata=df["ASKED YIELD"],   # y data
    p0=(0, 0, 0, 0, 0, 0, 0),      # initial value of the parameters
    maxfev=50000
)
```

```python
print(popt)
```

```python
df["MODEL FIT"] = f_model(df["DAYS TO MATURITY"], -1.49794283e+04, -2.79022361e-02,  2.13821665e-07, -1.96499447e+02, 
                          -7.06594731e+02,  1.38804175e+03, -4.86139722e+04)
ax = df.plot(
    x="DAYS TO MATURITY", y="ASKED YIELD",
    kind="line", linestyle="", marker="."
)
ax = df.plot(
    x="DAYS TO MATURITY", y="MODEL FIT",
    ax=ax, linewidth=1
)
```
